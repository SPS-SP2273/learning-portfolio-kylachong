**Exercise 1**

```python
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color="brown", linestyle="dashed",
             marker="o", markerfacecolor="yellow",
             ecolor="black",capsize=5,
             label="Y values with errors")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

**Exercise 2**

```python
# Generate x and y data for the function
x = np.linspace(0, 50, 500)
y = np.exp(-x / 10) * np.sin(x)

# Start plotting with the customizations
plt.figure(figsize=(10, 5))
plt.plot(x, y, label=r'$y = e^{-\frac{x}{10}}\sin(x)$', color='#1f77b4')

# Customizing the plot based on the requirements
plt.xlabel('x', fontsize=15, color='darkorange')
plt.ylabel('y', fontsize=15, color='darkorange', rotation=0)
plt.title(r'$y = e^{-\frac{x}{10}}\sin(x)$', fontsize=20)
plt.grid(True, alpha=0.25)
plt.xlim(0, 50)  # x range is already from 0 to 50
plt.ylim(-1, 1)  # Setting y limit as instructed

# Show the plot with the legend
plt.show()
```
